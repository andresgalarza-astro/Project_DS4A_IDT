# Carga base TM

```python
df_crimes = pd.read_csv(file_path2)
df_crimes
```

```python
df_crimes.info()
```

```python
for i in range(len(df_crimes['LATITUD_Y'])):
    df_crimes['LATITUD_Y'][i] = df_crimes['LATITUD_Y'][i].replace(",", ".")
```

```python
for i in range(len(df_crimes['LONGITUD_X'])):
    df_crimes['LONGITUD_X'][i] = df_crimes['LONGITUD_X'][i].replace(",", ".")
```

```python
df_crimes['LATITUD_Y'] = df_crimes['LATITUD_Y'].astype(float)
```

```python
df_crimes['LONGITUD_X'] = df_crimes['LONGITUD_X'].astype(float)
```

```python
df_crimes.info()
```

```python
gdf = gpd.GeoDataFrame(df_crimes, geometry=gpd.points_from_xy(df_crimes.LONGITUD_X, df_crimes.LATITUD_Y))
gdf
```

```python
gdf.info()
```

```python
base1 = gpd.sjoin(gdf,MapaLocalidades[['geometry','Nombre_Localidad','geometry2']], how='left',op='within')
base1
```

```python
base1.info()
```

```python
base1.geometry2.plot()
```
