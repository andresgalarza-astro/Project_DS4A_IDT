# Carga geojson

```python
file_path= 'D:/JUAN CAMILO/DS4A/Proyecto FInal/locashp/Loca.shp'
file_path2='D:/JUAN CAMILO/DS4A/Proyecto FInal/df_SITP.xlsx'
```
