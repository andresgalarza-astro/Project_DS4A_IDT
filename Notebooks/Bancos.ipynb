```python
from IPython.core.display import HTML
from bs4 import BeautifulSoup
from IPython.display import IFrame
import urllib # package required to interact with live webpage
import pandas as pd # will use to store the data from the webpage
```

```python
bank_key = ['bancamia','banco-av-villas','banco-caja-social','banco-corpbanca','banco-de-bogota','banco-de-occidente','banco-gnb-sudameris','banco-itau-1','banco-pichincha','banco-popular','bancolombia','bbva','scotiabank-colpatria']
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent}
bank_dict = {}
bank_dict['bank_key'] = []
bank_dict['bank_name'] = []
bank_dict['bank_address'] = []
bank_dict['bank_latitude'] = []
bank_dict['bank_longitude'] = []
counter = 0
hasData = False
for h in bank_key:
  counter = 1
  hasData = True
  while(hasData):
    site_url='https://www.civico.com/bogota/sucursales/'+h+'?page='+str(counter)
    #print(site_url)
    request=urllib.request.Request(site_url,None,headers)
    response = urllib.request.urlopen(request)
    site_content = response.read().decode('utf-8')
    soup = BeautifulSoup(site_content, 'html.parser')
    hasData = False
    for i in soup.find_all("div", {"class": "card-component place-card"}):
      hasData = True
      #print(h+','+str.strip(i['data-name'])+','+str.strip(i['data-address'])+','+i['data-latitude']+','+i['data-longitude'])
      bank_dict['bank_key'].append(h)
      bank_dict['bank_name'].append(str.strip(i['data-name']))
      bank_dict['bank_address'].append(str.strip(i['data-address']))
      bank_dict['bank_latitude'].append(i['data-latitude'])
      bank_dict['bank_longitude'].append(i['data-longitude'])
    if not hasData:
      break
    else:
      counter += 1
  
```

```python
df1 = pd.DataFrame(bank_dict)
df1#.style.hide_index()
```
