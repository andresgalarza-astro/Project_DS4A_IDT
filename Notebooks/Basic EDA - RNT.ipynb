```python
# Read Dataset
df=pd.read_excel('gdrive/My Drive/DS4A/Project/Datasets/5. Base de datos de establecimientos de alojamiento RNT.xlsx')
```

```python
# Drop records with NA values
df.dropna(inplace = True)
```

```python
# Final Dataset Description
df.info()
```

```python
df.head()
```

```python
df.describe().drop(columns = ['RNT', 'LATITUD', 'LONGITUD'])
```

```python
localidad_df = pd.DataFrame(df['LOCALIDAD'].value_counts()).reset_index().sort_values('LOCALIDAD')
ax = px.bar(localidad_df, x='LOCALIDAD', y='index', labels = {'LOCALIDAD':'# de Establecimientos de Turismo', 
                                                              'index':'Localidad'}, title = 'Establecimientos de Turismo por Localidad')
ax.show()
```

```python
barrio_df = pd.DataFrame(df['BARRIO'].value_counts()).reset_index().sort_values('BARRIO', ascending = False).head(10).sort_values('BARRIO')
ax = px.bar(barrio_df, x='BARRIO', y='index', labels = {'BARRIO':'# de Establecimientos de Turismo', 
                                                              'index':'Barrio'}, title = 'Establecimientos de Turismo por Barrio')
ax.show()
```

```python
categoria_df = pd.DataFrame(df['CATEGORIA'].value_counts()).reset_index().sort_values('CATEGORIA')
ax = px.bar(categoria_df, x='CATEGORIA', y='index', labels = {'CATEGORIA':'# de Establecimientos de Turismo', 
                                                              'index':'Categoría'}, title = 'Establecimientos de Turismo por categoría')
ax.show()
```

```python
hospedaje_df = df[(df['CANTIDAD_CAMAS'] > 0) & (df['CANTIDAD_HABITACIONES'] > 0)]
hospedaje_df
hospedaje_df.describe().drop(columns = ['RNT', 'LATITUD', 'LONGITUD'])
```

```python
localidad_df2 = pd.DataFrame(hospedaje_df['LOCALIDAD'].value_counts()).reset_index().sort_values('LOCALIDAD')
ax = px.bar(localidad_df2, x='LOCALIDAD', y='index', labels = {'LOCALIDAD':'# de Establecimientos de Turismo', 
                                                              'index':'Localidad'}, title = 'Establecimientos de Turismo por Localidad')
ax.show()
```

```python
barrio_df2 = pd.DataFrame(hospedaje_df['BARRIO'].value_counts()).reset_index().sort_values('BARRIO', ascending = False).head(10).sort_values('BARRIO')
ax = px.bar(barrio_df2, x='BARRIO', y='index', labels = {'BARRIO':'# de Establecimientos de Turismo', 
                                                              'index':'Barrio'}, title = 'Establecimientos de Turismo por Barrio')
ax.show()
```
