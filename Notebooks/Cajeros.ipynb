```python
from IPython.core.display import HTML
from bs4 import BeautifulSoup
from IPython.display import IFrame
import urllib # package required to interact with live webpage
import pandas as pd # will use to store the data from the webpage
import numpy as np
```

```python
atm_key = ['cajero-bancolombia','cajero-banco-caja-social','cajero-banco-de-bogota','cajero-banco-popular','cajeros-scotiabank-colpatria','cajeros-automaticos-davivienda','cajeros-banco-av-villas','cajeros-banco-de-occidente']
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent}
atm_dict = {}
atm_dict['atm_class'] = []
atm_dict['atm_name'] = []
atm_dict['atm_address'] = []
atm_dict['atm_latitude'] = []
atm_dict['atm_longitude'] = []
counter = 0
hasData = False
for h in atm_key:
  counter = 1
  hasData = True
  while(hasData):
    site_url='https://www.civico.com/bogota/sucursales/'+h+'?page='+str(counter)
    #print(site_url)
    request=urllib.request.Request(site_url,None,headers)
    response = urllib.request.urlopen(request)
    site_content = response.read().decode('utf-8')
    soup = BeautifulSoup(site_content, 'html.parser')
    hasData = False
    for i in soup.find_all("div", {"class": "card-component place-card"}):
      hasData = True
      #print(h+','+str.strip(i['data-name'])+','+str.strip(i['data-address'])+','+i['data-latitude']+','+i['data-longitude'])
      atm_dict['atm_class'].append(h)
      atm_dict['atm_name'].append(str.strip(i['data-name']))
      atm_dict['atm_address'].append(str.strip(i['data-address']))
      atm_dict['atm_latitude'].append(i['data-latitude'])
      atm_dict['atm_longitude'].append(i['data-longitude'])
    if not hasData:
      break
    else:
      counter += 1
  
```

```python
df1 = pd.DataFrame(atm_dict)
df1#.style.hide_index()
```
