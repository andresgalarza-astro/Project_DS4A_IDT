```python
from IPython.core.display import HTML
from bs4 import BeautifulSoup
from IPython.display import IFrame
import urllib # package required to interact with live webpage
import pandas as pd # will use to store the data from the webpage
import numpy as np
```

```python
diccionario = {}
diccionario['direccion'] = []
diccionario['localidad'] = []

for j in range(1,19):
  
  # leer la pagina
  site_url='https://www.bogotamiciudad.com/directoriodebogota/comercio/5-630/bogota/droguerias_y_farmacias?p='+str(j)+'#res'
  r = urllib.request.urlopen(site_url) # le hago la petición a la pagina web para leerla
  site_content = r.read().decode('utf-8') # códifico la información de la pagina web
  s = BeautifulSoup(site_content, 'html.parser')

  # función para extraer información
  for i in s.find_all('p')[3::2]:
    if type(i.text[3:].split('\r')) is list and len(i.text[3:].split('\r')) > 1:
      #print(k, i.text[3:].split('\r')[0].strip(), i.text[3:].split('\r')[1].strip())
      #k = k + 1
      diccionario['direccion'].append(i.text[3:].split('\r')[0].strip())
      diccionario['localidad'].append(i.text[3:].split('\r')[1].strip())
```

```python
df1 = pd.DataFrame(diccionario)
df1
```

```python
import requests
import lxml.html as html
import urllib3
import pandas as pd

diccionario2 = {}
diccionario2['nombre'] = []

xpath_par = "//div[@class='row list-row-even']"
xpath_impar = "//div[@class='row list-row-uneven']"
df = pd.DataFrame()

def leer_pagina(num, xp):
	url = "https://www.bogotamiciudad.com/directoriodebogota/comercio/5-630/bogota/droguerias_y_farmacias?p="+str(num)
	headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
	urllib3.disable_warnings()
	response = requests.get(url, headers=headers, verify=False)
	if response.status_code == 200:
		home = response.content.decode('utf-8')
		parsed = html.fromstring(home)
		
		listan = parsed.xpath(xp+"/h3/a/text()")
		for i in listan:
			diccionario2['nombre'].append(i)


i=1
total = 19
while i < total:
	#leer_pagina(i, xpath_par)
	leer_pagina(i, xpath_impar)
	i = i +1
    
i=1
while i < total:
	leer_pagina(i, xpath_par)
	#leer_pagina(i, xpath_impar)
	i = i +1

#df.info()
```
